```python
from clearml import Dataset

DATASET = "fashion_mnist"
# ["mnist", "cmnist", "fashion_mnist"]
```

___
# Upload a (new) Dataset

```python
ds = Dataset.create(
    dataset_name=DATASET,
    dataset_project="softtargets",
)
```

```python
ds.add_files(
    path=f"../data/{DATASET}_index.csv"
)
```

```python
import os
import pandas as pd
# Count the amount of samples per class

root_folder = "../data"
counts = []
labels = []

# Firstly, lets load the .csv file into pandas
df = pd.read_csv(os.path.join(root_folder, f"{DATASET}_index.csv"), sep=";", index_col=0)

# Then, lets get the amount of samples per class
counts = df['Class_Label'].value_counts()
labels = df['Class_Label'].unique()

ds.get_logger().report_histogram(
    title="Dataset Statistics",
    series="Class Counts",
    values=counts.sort_index().to_list(),
    xlabels=[str(x) for x in labels], #labels,
    xaxis="Class",
    yaxis="Count",
)
```

```python
ds.upload()
ds.finalize()
```

# Sync a new Version of an existing One


```python
parent_dataset_ids = ["...."] # List of IDs

ds_sync = Dataset.create(
    dataset_name=DATASET,
    dataset_project="softtargets",
    parent_datasets=[DATASET]
)
```

```python
# One could make changes to the datset without downloading it, like this:
ds_sync.remove_files()
ds.add_files()
```

```python
# But we have a local copy anyway, that represents how we want the dataset to look like
# in that case, it's easier to use the sync method
ds_sync.sync_folder(
    local_path=f"../data/{DATASET}_index.csv"
)
```

```python
ds_sync.finalize(
    auto_upload=True
)
```
